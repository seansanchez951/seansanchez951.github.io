```python
import numpy as np
import pandas as pd
import sqlite3
import math
import datetime
from datetime import date
```

```python
# This notebook is additional data cleaning to create a Bubble/Multi-line chart hybrid
# We will create a new csv of aggregated data from the 'genre-score-over-time' csv file
# We will take the average review score for each genre for each publication date
# This way each genre will only have one bubble per publication date to avoid vertical lines
# This aggregated data will be used for the multi-line chart
```

```python
dataset = pd.read_csv(r'C:\Users\seans\Desktop\CS 360 - Final Project\genre-score-over-time.csv')
```

```python
dataset
```

```python
new_df = dataset.groupby(['genre','pub_date'])['score'].mean().reset_index()
```

```python
new_df
```

```python
new_df.info()
```

```python
# create new column month extract 
# new_df['month'] = pd.DatetimeIndex(new_df['pub_date']).month
```

```python
new_df['year'] = new_df['pub_date'].str.split("-").str.get(0)
new_df['month'] = new_df['pub_date'].str.split("-").str.get(1)
new_df['day'] = new_df['pub_date'].str.split("-").str.get(2)
new_df['day'] = new_df['day'].str.split(" ").str.get(0)
```

```python
new_df['year'] = new_df['year'].astype(int)
new_df['month'] = new_df['month'].astype(int)
new_df['day'] = new_df['day'].astype(int)
```

```python
new_df.dtypes
```

```python
DATE = []

for x, y, z in zip(new_df.year, new_df.month, new_df.day):
    DATE.append(date(x, y, z))
```

```python
new_df['ymd_date'] = DATE
```

```python
new_df
```

```python
monthly_df = new_df.groupby(['genre','month','ymd_date'])['score'].mean().reset_index()
```

```python
monthly_df
```

```python
monthly_df.info()
```

```python
monthly_df['month_year'] = pd.to_datetime(monthly_df['ymd_date']).dt.to_period('M')
```

```python
monthly_df.head()
```

```python
new_df = monthly_df.groupby(['genre','month_year'])['score'].mean().reset_index()
```

```python
new_df
```

```python
new_df.to_csv('genre-avg-score-month-date.csv', header=True, index=False)
```
